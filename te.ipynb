```python
import pandas as pd
import os

# 1. Skapa data-mappen
os.makedirs("data", exist_ok=True)
print("Hämtar publik Spotify-data...")

# 2. Hämta datan direkt från en öppen GitHub-källa
url = "https://raw.githubusercontent.com/Bhasfe/ml-data/master/spotify-2023.csv"
df = pd.read_csv(url, encoding='ISO-8859-1')

print(f"Hittade fil med {df.shape[0]} rader och {df.shape[1]} kolumner.")

# 3. Banta ner datan så AI:n klarar av den
kolumner_att_behalla = [
    'track_name', 
    'artist(s)_name', 
    'released_year', 
    'streams', 
    'bpm', 
    'danceability_%'
]

df_clean = df[kolumner_att_behalla].head(100)

# 4. Spara ner filen i din data-mapp
spar_sokvag = "data/spotify_test.csv"
df_clean.to_csv(spar_sokvag, index=False, encoding='utf-8')

print(f"Färdigt! Den rensade filen är nu sparad på: {spar_sokvag}")
```
